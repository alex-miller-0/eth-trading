```python
def avg_close(df):
    return df["close"].mean()

def vol(df):
    return sum(df["volume"])


# For each row (minute), get the following columns pertaining to the last X minutes
def base_window_func(df, position, delay, func, new_col):
    
    # Get the subset of the df between the current position and the number of minutes we are looking back to
    sub_df = df.iloc[position-delay: position]

    # Map the function to a subset of the dataframe
    return {'date': df.iloc[position]["date"], new_col: func(sub_df)}



# Apply the base_window_func with the given delay
def apply_delay(df, delay, func, new_col):
    
    iloc_range = range(delay, len(df))
    delays = list(map(lambda x: base_window_func(df, x, delay, func, new_col), iloc_range))
    outside_range = list(map(lambda x: {'date': df.iloc[x]["date"], new_col: None}, range(0, delay)))
    new_list = delays + outside_range
    
    new_df = pd.DataFrame(new_list)

    return pd.merge(df, new_df, on='date')
```

```python
# I want to add the 24 hour average close
day_ago = 12    # Given 2 hour window, one day ago is 12 windows ago
two_day_ago = 24
df = apply_delay(df, day_ago, avg_close, "24_hr_avg")
df = apply_delay(df, two_day_ago, avg_close, "48_hr_avg")
df = apply_delay(df, day_ago, vol, "24_hr_vol")
df = apply_delay(df, two_day_ago, vol, "48_hr_vol")
df["above_avg"] = df.apply(lambda x: 1 if float(x["close"]) > float(x["24_hr_avg"]) else 0, axis=1)
```

Now I will use a more sophisticated approach, which will attempt to "predict" the price in the next period. For this, I want longer periods. ### Simulate trading
Now I am going to simulate a trading bot that follows the following rules:

* Buy if last X closes have been above the weighted average and the previous Y were below it
* Sell if the last X closes have been below the weighted average and the previous Y were above it

I will play with X and Y, but definitely I want Y to be larger than X and I probably want X to be 3 (=45 minutes). Y will probably be around 12 (3 hours)

```python
class TradeBot():
    def __init__(self):
        self.holding_buy = None     # (bool) if False, I am holding a sell
        self.next_y = None
        self.historic_X = pd.DataFrame()
        self.historic_y = list()
        self.model = None
        self.eth = 0
        self.btc = 0
        self.buys = 0
        self.sells = 0
        self.window_len = 7200      # In seconds --> 2 hrs
        self.buy_dates = list()
        self.sell_dates = list()
        
    # Initial setup
    def set_initial_params(self, starting_price):
        self.eth = 200
        self.start_price = starting_price
        self.btc = 200*starting_price
        self.buys = 0
        self.sells = 0
        self.holding_buy = True    
    
    # Train the model with the historic data in the instance
    def set_model(self):
        rfr = RandomForestRegressor()
        # Assign heavier weights to more recent data
        limit = len(self.historic_y)**2.2
        weights = list(map(lambda x: x**2.2/limit, range(len(self.historic_y))))
        
        # Fit and stash model
        rfr.fit(self.historic_X, self.historic_y, sample_weight=weights)
        self.model = rfr
        
    # Predict y given a new window of data
    def predict(self, new_window):
        y_pred = self.model.predict(new_window)
        return y_pred
    
    # Add a window of data to the instance
    def add_data(self, X, y):
        self.historic_X = self.historic_X.append(X)
        self.historic_y.append(y)

    # Sell 20% of my BTC worth of ETH
    def exec_sell(self, price):
        eth_to_sell = 0.2*self.btc / price
        self.btc += (eth_to_sell * price * 0.99)
        self.eth -= (eth_to_sell * 0.99)
    
    # Buy ETH equivalent to 20% of my BTC
    def exec_buy(self, price):
        eth_to_buy = 0.2*self.btc / price
        self.btc -= (eth_to_buy * price * 0.99)
        self.eth += (eth_to_buy * 0.99) 
        
    # Check if a trade should be made given the CURRENT window
    # @returns buy (bool), sell (bool)
    def inquire_trade(self, new_window):
        y_pred = self.predict(new_window)
        two_day_avg = self.historic_X.iloc[-1]["48_hr_avg"]
        one_day_avg = self.historic_X.iloc[-1]["24_hr_avg"]
        current_price = self.historic_X.iloc[-1]["open"]
        
        avg_recent = sum(self.historic_y[-4:])/4
        
        # Buy if previous close < 48 hr avg and predicted price > 48 hr avg
        if avg_recent < one_day_avg and y_pred > one_day_avg and self.holding_buy == False:
            print("(%.3f ETH, %.3f BTC) BUY: %.3f ETH @ %s"%(self.eth, self.btc, 0.2*self.btc/current_price, current_price))
            self.exec_buy(current_price)
            self.holding_buy = True
            self.buys += 1
            self.buy_dates.append(current_price)
            self.sell_dates.append(0)
            
        # Sell if previous close > 48 hr avg and predicted price < 48 hr avg
        elif avg_recent > one_day_avg and y_pred < one_day_avg and self.holding_buy == True:
            print("(%.3f ETH, %.3f BTC) SELL: %.3f ETH @ %s"%(self.eth, self.btc, 0.2*self.btc/current_price, current_price))
            self.exec_sell(current_price)
            self.holding_buy = False
            self.sells += 1
            self.sell_dates.append(current_price)
            self.buy_dates.append(0)
        
        else:
            self.sell_dates.append(0)
            self.buy_dates.append(0)
            
```

```python
start = len(y)//2                        # Start halfway through
start_price = df.iloc[start]["close"]
bot = TradeBot()
bot.set_initial_params(start_price)

start_eth = bot.eth
start_btc = bot.btc

# Set initial data
for i in range(start):
    bot.add_data(X.iloc[i], y.iloc[i])
```

```python
# Start running the model on the 301st data point (arbitrary buffer)
for i in range(start+1, len(y)):
    bot.add_data(X.iloc[i-1], y.iloc[i-1])
    bot.set_model()
    bot.inquire_trade(X.iloc[i])
```

```python
DAY_AGO = 96    # Given 15 minute window, one day ago is 96 windows ago

def check_close_below_avg(x):
    return False if x["close"] > 0.99*x["24_hr_avg"] else None
def check_close_above_avg(x):
    return False if x["close"] < 1.01*x["24_hr_avg"] else None

def check_vol_above(x):
    #return False if x["volume"] < 0.2*(x["24_hr_vol"]/DAY_AGO) else None
    return None





## TODO add condition that allows buy or sell if the MAJORITY of the lagging points were appropriately
#    above/below the 24_hr_avg

CUTOFF = 5    # The number of consecutive recent periods that will need to be above/below the 24_hr_avg 
              # to trigger a buy/sell
LOOKBACK = 30 # The number of periods we will be scanning backward when assessing a buy/sell

# Shave off the most recent periods
def cut_df(df, index):
    return df[index-LOOKBACK:index]


def check_trigger(df):
    
    # The number of consecutive periods below avg
    size = len(df)
    n_above = sum(map(lambda i: i, df[:-CUTOFF]["above_avg"]))
    n_below = size - CUTOFF - n_above   
    
     # The ratio above avg
    below_ratio = n_below/(n_above+n_below)
    above_ratio = 1-below_ratio
    
    # The number in the last X periods above avg
    last_periods_above = sum(map(lambda i: i, df[-CUTOFF:]["above_avg"]))
    last_periods_below = CUTOFF - last_periods_above
        
    if below_ratio > 0.8 and last_periods_above == CUTOFF:
        # Trigger buy
        return True, False
    elif above_ratio > 0.8 and last_periods_below == CUTOFF:
        # Trigger sell
        return False, True
    else:
        # Trigger nothing
        return False, False

# Buy eth with 10% of your btc
def buy_eth(eth, btc, price):
    btc_to_sell = 0.2*btc
    btc -= (btc_to_sell * 0.99)
    eth += (btc_to_sell / price * 0.99) 
    return eth, btc

# Sell 10% of your eth
def sell_eth(eth, btc, price):
    eth_to_sell = 0.2*eth
    btc += (eth_to_sell * price * 0.99)
    eth -= (eth_to_sell * 0.99)
    return eth, btc
```

```python
start = len(df)//2   # Start halfway through

eth = 100
start_price = df.iloc[start]["close"]
btc = 100*start_price
buys = 0
sells = 0
holding_buy = True
holding_sell = False

eth_holdings = [eth]
btc_holdings = [btc]
prices = [df.iloc[start]["close"]]
buy_dates = []
sell_dates = []

print("start ETH: %s, start BTC: %s"%(eth, btc))
for i in range(start, len(df)):
    
    buy, sell = check_trigger(cut_df(df, i))
    
    if buy and holding_sell == True:
        buys += 1
        eth, btc = buy_eth(eth, btc, df.iloc[i]["close"])
        eth_holdings.append(eth)
        btc_holdings.append(btc)
        prices.append(df.iloc[i]["close"])
        buy_dates.append(df.iloc[i]["date"])
        holding_sell = False; holding_buy = True
        
        
    elif sell and holding_buy == True:
        sells += 1
        eth, btc = sell_eth(eth, btc, df.iloc[i]["close"])
        eth_holdings.append(eth)
        btc_holdings.append(btc)
        prices.append(df.iloc[i]["close"])
        sell_dates.append(df.iloc[i]["date"])
        holding_sell = True; holding_buy = False

    
print("end ETH: %s, end BTC: %s"%(eth, btc))
print("%s buys, %s sells"%(buys, sells))
eth=100
btc=2
```
