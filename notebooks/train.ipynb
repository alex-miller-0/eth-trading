# Predict ETH
The overall goal is to predict if the price of ETH (in terms of BTC) is going to go up or down.

## Problem Definition
Specifically, I want to determine if the price of ETH is going to be **>1% higher in the next trading window**. Since ETH is traded 24/7, there are no true "open" or "close" times, so I want to chunk time windows into 6 hour chunks, thus I will define 4 trading periods a day:

* **00:00:00 GMT - 05:59:59 GMT**
* **06:00:00 GMT - 11:59:59 GMT**
* **12:00:00 GMT - 17:59:59 GMT**
* **18:00:00 GMT - 23:59:59 GMT**

## Price Definition
The price will be defined as the **last trade on the first minute** of any given trading window, as pulled from BitMEX.

Okay, let's get started...

```python
import pandas as pd
import numpy as np
import pymysql as sql
import sys,os
from datetime import datetime, timedelta
```

```python
host = os.environ['ETH_SQL_HOST']
password = os.environ['ETH_SQL_PASSWORD']
user = os.environ['ETH_SQL_USER']
db = os.environ['ETH_SQL_DB']
con = sql.connect(host=host, user=user, passwd=password, db=db)
```

```python
df_prices = pd.read_sql('SELECT * FROM BitMexPrice ORDER BY timestamp', con)
df_trades = pd.read_sql('SELECT * FROM BitMexTrades ORDER BY timestamp', con)
```

```python
df = pd.merge(df_prices, df_trades, on='timestamp')
del df["id_x"]
del df["id_y"]
```

### Transformations
I need to transform the timestamp into two numbers:

* **window** (window 0 is 2015-08-10 00:00:00 - 2016-08-10 06:00:00)
* **window_minute* (there are 6x60 = 360 total minutes in the window)

I also want the minute since the starting point, which I will simply call **minute**.

```python
def get_window(ts):
    origin = datetime(2015,8,10,0,0,0)
    seconds = (ts - origin).total_seconds()
    return seconds // 21600

def get_window_minute(ts):
    hour = ts.hour
    minute = ts.minute
    return ((hour*60) + minute) - ((hour//6)*6*60)

def get_minute(ts):
    return 360*get_window(ts) + get_window_minute(ts)
```

```python
df["window"] = df["timestamp"].apply(get_window)
df["window_minute"] = df["timestamp"].apply(get_window_minute)
df["minute"] = df["timestamp"].apply(get_minute)
del df["timestamp"]
```

### Features
Here I will list a bunch of features that I want to include. Remember that the time units are 6 hour windows.

* Volume traded in window
* Net price change in window
* Max price in window
* Min price in window
* Average price in window
* Standard deviation of price in window
* Number of trades in window
* Number of upward swings
* Number of downward swings
* Max consecutive upward swings
* Max consecutive downward swings

```python
# Get volume traded in window
def get_window_volume(df):
    return sum(df["volume"])

# Get net price change in window (i.e. last price - first price)
def get_window_net(df):
    delta = df.iloc[len(df)-1]["price"] - df.iloc[0]["price"]
    return delta

# Max price in window
def get_max_window_price(df):
    return max(df["price"]) 

# Min price in window
def get_min_window_price(df):
    return min(df["price"])

# Average window price
def get_avg_window_price(df):
    return df["price"].mean()

# Standard deviation of window price
def get_std_window_price(df):
    return df["price"].std()

# Number of trades in window
def get_num_trades(df):
    return sum(df["trades"])

# Number of upward swings
def get_num_up_swings(df):
    return sum(map(lambda i: int(df.iloc[i]["price"] > df.iloc[i-1]["price"]), range(1, len(df))))

# Number of downward swings
def get_num_down_swings(df):
    return sum(map(lambda i: int(df.iloc[i]["price"] < df.iloc[i-1]["price"]), range(1, len(df))))

# Max consecutive upward swings
def get_max_up_swings(df):
    max_swing = 0
    current_swing = 0
    for i in range(1, len(df)):
        if df.iloc[i]["price"] > df.iloc[i-1]["price"]:
            current_swing += 1
        elif df.iloc[i]["price"] < df.iloc[i-1]["price"]:
            if current_swing > max_swing:
                max_swing = current_swing
                current_swing = 0
        else:
            pass
    return max_swing

# Max consecutive downward swings
def get_max_down_swings(df):
    max_swing = 0
    current_swing = 0
    
    for i in range(1, len(df)):
        
        if df.iloc[i]["price"] < df.iloc[i-1]["price"]:
            current_swing += 1
            if current_swing > max_swing:
                max_swing = current_swing
            
        elif df.iloc[i]["price"] > df.iloc[i-1]["price"]:
            if current_swing > max_swing:
                max_swing = current_swing
            current_swing = 0
        
        else:
            pass
    
    return max_swing
```

```python
# For each row (minute), get the following columns pertaining to the last X minutes
def base_window_func(df, position, minutes, func, new_col):
    
    # Get the subset of the df between the current position and the number of minutes we are looking back to
    sub_df = df.iloc[position-minutes: position]

    # Map the function to a subset of the dataframe
    return {'minute': df.iloc[position]["minute"], new_col: func(sub_df)}


# Apply the base_window_func with the given delay
def apply_delay(df, delay, func, new_col):
    
    iloc_range = range(delay, len(df))
    delays = list(map(lambda x: base_window_func(df, x, delay, func, new_col), iloc_range))
    outside_range = list(map(lambda x: {'minute': df.iloc[x]["minute"], new_col: None}, range(0, delay)))
    new_list = delays + outside_range
    
    print("len_delays: %s"%len(delays))
    print("len_new_list: %s"%len(new_list))
    new_df = pd.DataFrame(new_list)

    return pd.merge(df, new_df, on='minute')
```

```python
# Get all of the delay columns given a delay
def get_delay_cols(df, delay, prefix):
    df = apply_delay(df, delay, get_window_net, "%s_del"%prefix)
    df = apply_delay(df, delay, get_window_volume, "%s_vol"%prefix)
    df = apply_delay(df, delay, get_max_window_price, "%s_max_price"%prefix)
    
    df = apply_delay(df, delay, get_min_window_price, "%s_min_price"%prefix)
    df = apply_delay(df, delay, get_avg_window_price, "%s_avg_price"%prefix)
    df = apply_delay(df, delay, get_std_window_price, "%s_std_price"%prefix)
    df = apply_delay(df, delay, get_num_trades, "%s_trades"%prefix)
    df = apply_delay(df, delay, get_num_up_swings, "%s_up_swings"%prefix)
    df = apply_delay(df, delay, get_num_down_swings, "%s_down_swings"%prefix)
    df = apply_delay(df, delay, get_max_up_swings, "%s_max_consec_up_swings"%prefix)
    df = apply_delay(df, delay, get_max_down_swings, "%s_max_consec_down_swings"%prefix)
    
    return df

def pipeline(df):
    df = get_delay_cols(df, 60, "6h")
    #df = get_delay_cols(df, 720, "12h")
    #df = get_delay_cols(df, 1440, "24h")
    return df
```
